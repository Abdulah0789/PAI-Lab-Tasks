```python
numerical_cols = df.select_dtypes(include=['int64']).columns.tolist()
float_cols = df.select_dtypes(include=['float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Numerical Columns (Integer):", numerical_cols)
print("Float Columns:", float_cols)
print("Categorical Columns:", categorical_cols)
```

```python
def fill_missing_values_by_dtype(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype in ['float64', 'float32']:
                df[col] = df[col].fillna(df[col].mean())
            elif df[col].dtype in ['int64', 'int32']:
                df[col] = df[col].fillna(int(df[col].mean()))
            elif df[col].dtype == 'object':
                mode_value = df[col].mode()
                if not mode_value.empty:
                    df[col] = df[col].fillna(mode_value.iloc[0])
                else:
                    df[col] = df[col].fillna("Unknown")

fill_missing_values_by_dtype(df)
```

```python
def convertFloatintoInt(cols):
    for col in cols:
        df[col] = df[col].astype('int64') 
columns = df.select_dtypes(include=['float']).columns  
convertFloatintoInt(columns)
```

```python
def dataEncoder(cols):
    for i in cols:
        dataLabelEncoder = LabelEncoder()
        df[i] = dataLabelEncoder.fit_transform(df[i])
columns = [catgory for catgory in df.columns if df[catgory].dtype == 'object']
dataEncoder(columns)
```
